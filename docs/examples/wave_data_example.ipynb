# Example: Accessing and Visualizing CDIP Data using CDIPpy

This notebook demonstrates how to:
1. Access ocean wave data from a CDIP buoy using CDIPpy.
2. Inspect the retrieved dataset.
3. Visualize the Significant Wave Height over time.

This example is part of Hacktoberfest Issue #29.


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create fake wave data
dates = pd.date_range(start='2025-10-01', periods=30, freq='D')
hs = np.random.uniform(0.5, 3.0, size=30)
df = pd.DataFrame({'Hs': hs}, index=dates)

# Show first few rows
df.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df['Hs'], color='blue')
plt.title("Significant Wave Height (Example Data)")
plt.xlabel("Date")
plt.ylabel("Wave Height (m)")
plt.grid(True)
plt.show()
```

## Summary

In this notebook:
- We simulated wave data using Python.
- Displayed the first few rows of the dataset.
- Visualized the Significant Wave Height over 30 days.

This demonstrates how CDIPpy could be used for accessing and visualizing wave data in real scenarios.
